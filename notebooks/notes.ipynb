### Github_01
```
### 🔑 1. トークンの発行手順（ブラウザでの操作）

GitHubにログインした状態で、以下の順にクリックしてください：

1. 右上の自分のアイコン → **Settings**
2. 左サイドメニューの一番下 → **Developer settings**
3. **Personal access tokens** → **Tokens (classic)** を選択
4. **Generate new token** → **Generate new token (classic)** をクリック
5. **Note**: 「For Laptop」など適当な名前を付ける
6. **Expiration**: 「No expiration」または長めの期間にする
7. **Select scopes**: 一番上の **`repo`** にチェックを入れる（これで十分です）
8. 一番下の **Generate token** をクリック
```

### Google Gemini_01 magic command
```
① 基本的な質問（テキスト出力）
セルの先頭に %%ai gemini-2.5-flash （または使いたいモデル名）を指定し、2行目以降にプロンプトを書きます。

Markdown
%%ai gemini-2.5-flash
Pythonでクイックソートを実装するコードを書いて、簡単な解説も付けてください。
② 出力フォーマットを指定する (-f)
デフォルトではプレーンテキストで返ってきますが、-f markdown や -f html をつけることで、数式やグラフ、綺麗な装飾付きのMarkdownウェブ表示として結果を受け取ることができます。

Markdown
%%ai gemini-2.5-flash -f markdown
データサイエンスのロードマップを、見出しや箇条書きを使って綺麗にまとめて。
③ エラーの理由を聞く（直前のセルの参照）
プログラムを実行してエラーが出た際、わざわざコピー＆ペーストしなくても、直前のエラー内容を自動でGeminiに渡して解説させることができます。

Markdown
%%ai gemini-2.5-flash
なぜ上のセルでエラーが出たのか、原因と修正コードを教えて。
```
