# youtube download

```python
import subprocess
from pathlib import Path
import os

# 1. Display an interactive popup box to input the URL
url = input("📥 Enter YouTube URL: ").strip()

if not url:
    print("❌ Error: No URL provided. Process aborted.")
else:
    # 2. Resolve the script path (automatically expands '~')
    script_path = Path("~/youtube/interactive_dl.sh").expanduser()

    print(f"🎬 Starting download... \nURL: {url}\n")

    # 3. Execute the shell script with the URL as an argument
    # Using shell=True ensures your current virtual environment path is preserved
    result = subprocess.run(
        f"bash {script_path} '{url}'", shell=True, capture_output=True, text=True
    )

    # 4. Display the execution logs
    print("--- Execution Log ---")
    print(result.stdout)

    if result.stderr:
        print("--- Standard Error / Warnings ---")
        print(result.stderr)
```

```python
import time
import switch_subtitle

# 初期設定（Trueなら最初は日本語、Falseなら英語からスタート）
is_japanese = True

print("Jupyter上で字幕の切り替えループを開始します。")
print("停止するには、上部の『■ (Interrupt the kernel)』ボタンを押してください。\n")

try:
    while True:
        # 言語の決定
        lang = "ja" if is_japanese else "en"
        
        # 実行ログの出力
        current_time = time.strftime('%H:%M:%S')
        print(f"[{current_time}] 字幕を '{lang}' に切り替えました。")
        
        # 切り替え関数の実行
        switch_subtitle.switch_subtitles(lang)
        
        # フラグを反転（次回の言語を切り替える）
        is_japanese = not is_japanese
        
        # 3分（180秒）待機
        time.sleep(30)

except KeyboardInterrupt:
    print("\n[STOP] カーネルが中断されました。ループを停止します。")
```
