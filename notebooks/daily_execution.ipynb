```python
import subprocess

def run_commands_fixed(cmd_string):
    process = subprocess.Popen(
        cmd_string,  # ← 'cmd_str' から 'cmd_string' に修正
        shell=True,
        executable='/bin/bash',
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT, 
        text=True
    )
    
    for line in process.stdout:
        print(line, end="")
        
    process.wait()
```

# Date

```python
from datetime import datetime

# Get today's date
today = datetime.now()

# Get the week number starting from Sunday
# %U: Week number of the year (Sunday as the first day of the week)
week_number = today.strftime("%U")

print(f"Today's date: {today.strftime('%Y-%m-%d')}")
print(f"Week Number (Sunday start): {week_number}")
```
