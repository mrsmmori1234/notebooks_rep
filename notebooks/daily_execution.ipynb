# daily check

```python
import calendar
from datetime import datetime

# --------------------------------------------------------------------
# 1. SETUP BASE DATE INFO
# --------------------------------------------------------------------
today = datetime.now()
base_year = today.year
base_month = today.month
current_day = today.day  # Used to pinpoint exactly "Today"

# --------------------------------------------------------------------
# 2. DEFINE NEON COLOR ESCAPE CODES
# --------------------------------------------------------------------
STYLE_H3 = "\033[1m"
RESET = "\033[0m"

COLOR_TODAY = "\033[48;5;197m\033[38;5;232m\033[1m"  # Glowing Hot Pink Box
COLOR_SG    = "\033[38;5;51m\033[1m"                # Cyan Bold Text
COLOR_JP    = "\033[38;5;226m\033[1m"               # Yellow Bold Text
COLOR_BOTH  = "\033[48;5;99m\033[38;5;255m\033[1m"   # Purple BG (Overlap)

# --------------------------------------------------------------------
# 3. PRINT GENERATED CALENDARS FOR THE NEXT 3 MONTHS (4 MONTHS TOTAL)
# --------------------------------------------------------------------
cal = calendar.Calendar(calendar.SUNDAY)

# Loop 4 times: current month + 3 months ahead
for i in range(3):
    # Calculate target month and year dynamically to handle calendar rollovers smoothly
    target_month = base_month + i
    target_year = base_year
    
    if target_month > 12:
        target_month -= 12
        target_year += 1
        
    active_sg = []
    active_jp = []

    # Fetch holidays dynamically for the targeted month
    try:
        import holidays
        sg_package = holidays.country_holidays('SG', years=target_year)
        jp_package = holidays.country_holidays('JP', years=target_year)
        
        for date_obj in sg_package.keys():
            if date_obj.month == target_month:
                active_sg.append(date_obj.day)
                
        for date_obj in jp_package.keys():
            if date_obj.month == target_month:
                active_jp.append(date_obj.day)
    except (ImportError, Exception):
        # Fallback maps fallback automatically if package is not present
        sg_fallback = {
            1: [1, 30, 31], 3: [29], 4: [2], 5: [1, 27], 
            6: [15], 8: [9, 10], 11: [8], 12: [25]
        }
        jp_fallback = {
            1: [1, 2, 3, 12], 2: [11, 23], 3: [21], 4: [29], 
            5: [3, 4, 5, 6], 7: [20], 8: [11], 9: [21, 22, 23], 
            10: [12], 11: [3, 23], 12: [31]
        }
        # For target years outside fallback definitions, default to empty arrays
        if target_year == base_year:
            active_sg = sg_fallback.get(target_month, [])
            active_jp = jp_fallback.get(target_month, [])

    # Print Month Header layout
    month_name = calendar.month_name[target_month]
    print(f"\n{STYLE_H3}### CALENDAR FOR {month_name.upper()} {target_year} ###{RESET}")
    print(" Su  Mo  Tu  We  Th  Fr  Sa")

    # Generate the grid array
    weeks = cal.monthdayscalendar(target_year, target_month)

    for week in weeks:
        week_line = ""
        for day in week:
            if day == 0:
                week_line += "    "
            else:
                day_str = f"{day:2}"
                
                # Check if this exact day is "Today" (Only highlight matching current month and year)
                is_today = (day == current_day and target_month == base_month and target_year == base_year)
                is_sg = (day in active_sg)
                is_jp = (day in active_jp)
                
                if is_today:
                    week_line += f" {COLOR_TODAY}{day_str}{RESET}"
                elif is_sg and is_jp:
                    week_line += f" {COLOR_BOTH}{day_str}{RESET}"
                elif is_sg:
                    week_line += f" {COLOR_SG}{day_str}{RESET}"
                elif is_jp:
                    week_line += f" {COLOR_JP}{day_str}{RESET}"
                else:
                    week_line += f" {day_str} "
                    
        print(week_line)

# --------------------------------------------------------------------
# 4. SHARED LEGEND SYSTEM
# --------------------------------------------------------------------
print("\n" + "-" * 28)
print(f"{COLOR_TODAY} ID {RESET} Today")
print(f"{COLOR_SG} •• {RESET} Singapore Holiday")
print(f"{COLOR_JP} •• {RESET} Japan Holiday")
print("-" * 28 + "\n")
```
