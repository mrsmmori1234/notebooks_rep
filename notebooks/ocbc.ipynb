# OCBC 365 Account

```python
import pandas as pd
import glob
import os

# 1. すべての対象ファイルを取得

search_path = f"../Win_Downloads/TransactionHistory*.csv"
all_files = glob.glob(search_path)

li = []

for filename in all_files:
    temp_df = pd.read_csv(filename, encoding='cp932', skiprows=5, quotechar='"')
    temp_df['Transaction date'] = pd.to_datetime(temp_df['Transaction date'], dayfirst=True)
    temp_df['Description'] = temp_df['Description'].str.replace(r'\s+', ' ', regex=True).str.strip()
    li.append(temp_df)

df = pd.concat(li, axis=0, ignore_index=True)
df = df.drop_duplicates()
df = df.sort_values('Transaction date').reset_index(drop=True)
df = df.sort_values('Transaction date', ascending=False).reset_index(drop=True)
```

```python
df.to_csv('./outputs/ocbc.csv', index=False, encoding='utf-8-sig')
```
