```python
commands = [
    "source ~/.pyenv/versions/3.12.2/envs/youtube-env/bin/activate",
    "cd ~/youtube",
    #"./update.sh",
    "./channel_mp4.sh"
]
```

```python
import subprocess

current_cmd = ""
try:
    for cmd in commands:
        if current_cmd:
            current_cmd += " && " + cmd
        else:
            current_cmd = cmd
            
        print(f"▶ Running: {cmd}")
        
        result = subprocess.run(
            current_cmd,
            shell=True,
            executable="/bin/bash",
            capture_output=True,
            text=True
        )

        if result.stdout:
            print(result.stdout.strip())
            
        if result.stderr:
            print(f"[Error Output]:\n{result.stderr.strip()}")
            
        if result.returncode != 0:
            print(f"❌ Failed: Error Code {result.returncode} Stopped")
            break
            
except Exception as e:
    print(f"Unexpected Error: {e}")
```
